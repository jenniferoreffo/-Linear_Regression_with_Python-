```python
#importing necessary libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# column names and meaning
column_names = {'date':'Date','Appliances':'Appliances','lights':'lights','T1': 'Temperature in kitchen area'
                ,'RH_1':'Humidity in kitchen area','T2': 'Temperature in living room area'
                ,'RH_2': 'Humidity in living room area','T3':'Temperature in laundry room area'
                ,'RH_3':'Humidity in laundry room area','T4':'Temperature in office room','RH_4':'Humidity in office room'
                ,'T5':'Temperature in bathroom','RH_5':'Humidity in bathroom','T6':'Temperature outside the building'
                ,'RH_6':'Humidity outside the building','T7':'Temperature in ironing room','RH_7':'Humidity in ironing room'
                ,'T8':'Temperature in teenager room 2','RH_8':'Humidity in teenager room 2','T9':'Temperature in parents room'
                ,'RH_9':'Humidity in parents room','T_out':'Temperature outside','Press_mm_hg':'Pressure_mm_hg' 
                ,'RH_out': 'Humidity outside','Wind speed':'Wind speed','Visibility':'Visibility','Tdewpoint':'Tdewpoint'
                ,'rv1':'Random variable 1','rv2':'Random variable 2'}
energy_data = energy_data.rename(columns=column_names)
```

```python
energy_data.hist(bins=50, figsize=(20,15)) #a quick feel of the data using histogram
plt.show()
```

```python
# dropping the 'date' and 'lights' columns from the dataset as they are not needed
data1 = energy_data.drop(columns=['Date','lights'])
```

```python
data1.head()
```

```python

# Normalise the dataset using the min max scaler to a common scale
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_data1 = pd.DataFrame(scaler.fit_transform(data1), columns=data1.columns)
features_data1 = normalised_data1.drop(columns=['Appliances'])
appliances_target = normalised_data1['Appliances']
```

```python
features_data1.head()
```

```python
#Now, we split our dataset into the training(70%) and testing(30%) dataset. Remember we had divided the dataset into features and target variables.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_data1, appliances_target, test_size=0.3, random_state=42)
```

```python
# printing the values of the train and test set
print(len(x_train), "train +", len(y_train,), "test")
print(len(x_test), "train +", len(y_test,), "test")
```

```python
# next, we fit our Linear Regression model to the training dataset
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()

#fit the model to the training dataset
linear_model.fit(x_train, y_train)

#obtain predictions
predicted_values = linear_model.predict(x_test)
```

```python
# Residual Sum of Squares(RSS)
rss = np.sum(np.square(y_test - predicted_values))
print('Residual Sum of Squares(RSS):', round(rss, 2))
```

```python
# Root Mean Square Error(RMSE)
from sklearn.metrics import  mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
print('Root Mean Square Error(RMSE):', round(rmse, 3))
```

```python
# Calculate the R-squared
from sklearn.metrics import r2_score
r2_scor = r2_score(y_test, predicted_values)
print('Coefficient of Determination/R-squared:', round(r2_scor, 2))
```

```python
#Calculate the Mean Absolute Error(MAE)
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
print('Mean Absolute Error(MAE):', round(mae,2))
```

```python
# this function returns the weight of every feature
def get_weights_df(model, feat, col_name):
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df

linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
```

```python
#calling the function on the dataset
linear_model_weights
```

```python
# Feature with the lowest weight
linear_model_weights.iloc[linear_model_weights['Linear_Model_Weight'].idxmin()]
```

```python
# Feature with the highest weight
linear_model_weights.iloc[linear_model_weights['Linear_Model_Weight'].idxmax()]
```

```python
# fit Lasso Regression into the train set 
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
lasso_pred = lasso_reg.predict(x_test)

# calculate the lasso feature weights using the get_weights_df function
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')
lasso_weights_df
```

```python

# total numbet of non-zero lasso feature weights
print('Number of non-zero feature weights:', (lasso_weights_df['Lasso_weight'] != 0).sum())
```

```python
# lasso regression Root Mean Square Error(RMSE)
lasso_rmse = np.sqrt(mean_squared_error(y_test, lasso_pred))
print('Root Mean Square Error(RMSE):', round(lasso_rmse, 3))
```

```python
#Ridge Regression
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)

ridge_pred = ridge_reg.predict(x_test)
```

```python
#Root Mean Square Error(RMSE)
ridge_rmse = np.sqrt(mean_squared_error(y_test, ridge_pred))
print('Root Mean Square Error(RMSE):', round(ridge_rmse, 3))
```
